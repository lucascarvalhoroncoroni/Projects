```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Actor's base
graph = nx.read_edgelist("./networks/actor.edgelist.txt")
```

```python
# Generating the graph sequence
degree_sequence = sorted((d for n, d in graph.degree()), reverse=True)
```

```python
# Generating the degree plot using a normal scale
plt.plot(degree_sequence, "b-", marker="o")
plt.title("Degree Rank Plot")
plt.ylabel("Degree")
plt.xlabel("Rank")
plt.show()
```

```python
# Generating the degree plot using a loglog scale
plt.loglog(degree_sequence, "b-", marker="o")
plt.title("Degree Rank Plot")
plt.ylabel("Degree")
plt.xlabel("Rank")
plt.show()
```
