### Consider a G(N,p) network with N = 3000 nodes, connected to each other with probability p = 1/1000

(a) What is the expected number of links, 〈L〉?

```python
N = 3000
p = 1/1000
# According to the formulas <k> = 2L/N and <k> = p * (N − 1) we have:
# p*(N-1) = 2*L/N
# So on average : 
L = (p * N * (N - 1))/2
L
```

(b) In which regime is this network?

There are four regimes: <br>
Subcritical, Critical, Supercritical, and, Connected

```python
# So, calculating <k> we can determine it's regime
avg_k = p * (N - 1)
avg_k # This is a supercritical network
```

(c) Plot the degree distribution p_k of this network (approximate with a Poisson degree distribution).

The poisson distribution uses the formula  : <br> pk = (e^-k) * (&lt;k&gt;^k)/k!<br>So, we have :

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# Building an array to represent the x_axis, in this case, k
# Since L is the average number of Links, it does not make sence to have a k bigger than L
x_axis = np.arange(0, L, 1)

# mu is <k>
# loc is where the sequence starts, moves the graph to tthe right
y_axis = poisson.pmf(x_axis, mu=avg_k, loc=0)
```

```python
# Showing the distribution
plt.plot(x_axis, y_axis)
plt.show()
```

The graph does not look good, because, this shows the distribution of the probability (y axys) of a random node having a degree k (x axys).<br>
Since the average degree is 2.999, a radom node has a higher probability of having a degree 2.999 <br>
<b>This probability is diferent than the one used before<b>

```python
# Building a better view
x_axis = np.arange(0, 10, 1)
y_axis = poisson.pmf(x_axis, mu=avg_k, loc=0)
```

```python
# Showing the distribution
plt.plot(x_axis, y_axis)
plt.show()
```

(d) Given the number of nodes N=3000, calculate the probability p_c so that the network is at the critical point.

```python
# <k> must be 1, so:
avg_k = 1
p = avg_k / (N - 1)
p
```
