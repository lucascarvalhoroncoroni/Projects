```python
import csv
```

```python
nodes = []
links = {}
```

```python
with open('nodes.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        nodes.append(row[0])
```

```python
with open('links.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        if row[0] in links:
            links[row[0]].append(row[1])
        else:
            links[row[0]] = [row[1]]
```
